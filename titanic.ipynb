```python
# read in titanic dataset (train.csv) that i found at this website: https://www.kaggle.com/c/titanic/data
dataFrame = pd.read_csv("train.csv")
#display the column headings and  first 6 rows
dataFrame.head()
```

```python
#reverse order of columns
columns = dataFrame.columns.tolist()
columns = columns[::-1]
dataFrameReversed = dataFrame[columns]
#display the column headings and  first 6 rows that have now be reversed
dataFrameReversed.head()
```

```python
#write reversed data to a file
export_csv = dataFrameReversed.to_csv("reversedColumns.csv")
```

```python
#drop alternating columns from dataFrame
dataFrameAlternates=dataFrame[dataFrame.columns[::2]]
#display headings of dataFrame columns
dataFrameAlternates.head()
```

```python
#write data to a file
export_csv = dataFrameAlternates.to_csv("alternateColumns.csv")
```
